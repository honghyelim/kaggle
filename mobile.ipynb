### (2) 데이터 스케일링(Data Scaling)

데이터 스케일링의 방법으로 총 네가지를 시도하여 성과를 svm으로 측정한 결과, Standard Scaler 가 가장 높게 나왔으므로
이하 분석에서는 Standard Scaler를 기준으로 분석을 수행 

```python
x = df.drop(['price_range'], axis=1).values # 독립변인들의 value값만 추출
y = df['price_range'].values # 종속변인 추출
```

```python
#(1) StandardScaler
# 각 feature의 평균을 0, 분산을 1로 변경합니다. 모든 특성들이 같은 스케일을 갖게 됩니다.
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler  # 표준화 패키지 라이브러리 
x = StandardScaler().fit_transform(x) # x객체에 x를 표준화한 데이터를 저장

features = ['battery_power', 'blue', 'clock_speed', 'dual_sim','fc','four_g','int_memory','m_dep','mobile_wt',
            'n_cores','pc','px_height','px_width','ram','sc_h','sc_w','talk_time','three_g ','touch_screen','wifi']
pd.DataFrame(x, columns=features).head()


from sklearn.svm import SVC
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, test_size=0.3)
svc = SVC()
svc.fit(x_train, y_train)
print('StandardScaler_test accuracy: %.3f'%(svc.score(x_test, y_test)))
```

```python
#(2) RobustScaler
#모든 특성들이 같은 크기를 갖는다는 점에서 StandardScaler와 비슷하지만, 평균과 분산 대신 median과 quartile을 사용합니다. RobustScaler는 이상치에 영향을 받지 않습니다.


from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
x_train_scale = scaler.fit_transform(x_train)

from sklearn.svm import SVC
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, test_size=0.3)
svc = SVC()
svc.fit(x_train, y_train)
print('RobustScaler_test accuracy: %.3f'%(svc.score(x_test, y_test)))
```

```python
#(3) MinMaxScaler
#모든 feature가 0과 1사이에 위치하게 만듭니다. 데이터가 2차원 셋일 경우, 모든 데이터는 x축의 0과 1 사이에, y축의 0과 1사이에 위치하게 됩니다.
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train_scale = scaler.fit_transform(x_train)

from sklearn.svm import SVC
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, test_size=0.3)
svc = SVC()
svc.fit(x_train, y_train)
print('MinMaxScaler_test accuracy: %.3f'%(svc.score(x_test, y_test)))
```

```python
#(4) Normalizer
#StandardScaler, RobustScaler, MinMaxScaler가 각 columns의 통계치를 이용한다면 Normalizer는 row마다 각각 정규화됩니다. Normalizer는 유클리드 거리가 1이 되도록 데이터를 조정합니다. (유클리드 거리는 두 점 사이의 거리를 계산할 때 쓰는 방법, L2 Distance)

from sklearn.preprocessing import Normalizer
scaler = Normalizer()
x_train_scale = scaler.fit_transform(x_train)

from sklearn.svm import SVC
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, test_size=0.3)
svc = SVC()
svc.fit(x_train, y_train)
print('Normalizer_test accuracy: %.3f'%(svc.score(x_test, y_test)))
```

* 스케일링 결과 : Standard Scaler 을 썼을 경우의 SVM 성능이 0.88로 가장 높다 .

## 릿지회귀(Ridge) (정확도: 92%)

```python

import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge 
from sklearn.model_selection import train_test_split

x = df[['battery_power',
        'blue',
        'clock_speed',
        'dual_sim',
        'fc','four_g',
        'int_memory',
        'm_dep',
        'mobile_wt',
        'n_cores',
        'pc','px_height',
        'px_width',
        'ram',
        'sc_h',
        'sc_w',
        'talk_time',
        'three_g',
        'touch_screen',
        'wifi']]
y = df[['price_range']]


x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, test_size=0.3) # 데이터셋을 훈련셋과 테스트셋으로 분리
rdg = Ridge()
rdg.fit(x_train, y_train) # 릿지 회귀 모델 훈련
```

```python
print('훈련 세트 점수 : {:.3f}'.format(rdg.score(x_train, y_train)))
print('테스트 세트 점수 : {:.3f}'.format(rdg.score(x_test, y_test)))
```

```python
# 여기서 alpha값 0.1, 10 으로 시도 해보겠다

#alpha 낮게 - 규제 약하게 - 데이터에 더 많이 피팅 - 과대적합 방향
#alpha 높게 - 규제 강하게 - 모든 계수를 0 에 가깝게 - 과소적합 방향
```

```python
#alpha = 0.1
rdg01 = Ridge(alpha = 0.1)
rdg01.fit(x_train, y_train) # 릿지 회귀 모델 훈련
print('훈련 세트 점수 : {:.3f}'.format(rdg01.score(x_train, y_train)))
print('테스트 세트 점수 : {:.3f}'.format(rdg01.score(x_test, y_test)))
```

```python
#alpha = 10
rdg01 = Ridge(alpha = 10)
rdg01.fit(x_train, y_train) # 릿지 회귀 모델 훈련
print('훈련 세트 점수 : {:.3f}'.format(rdg01.score(x_train, y_train)))
print('테스트 세트 점수 : {:.3f}'.format(rdg01.score(x_test, y_test)))
```

## 로지스틱 회귀(정확도 : 95%)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
#Divide the dataset into X and y
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
#scale the value of X
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)

#로지스틱 회귀 수행 
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0, multi_class = 'auto', solver = 'lbfgs')
classifier.fit(X_train, y_train)

```

```python
#predicting the values
y_pred = classifier.predict(X_test)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
cm_Log = confusion_matrix(y_test, y_pred)
cm_Log
```

```python
print(classification_report(y_test, y_pred))
```

```python
sns.heatmap(cm_Log, annot=True, fmt = 'g')
```

* 결과
- 정밀도(Precision) 95%
- 재현율(Recall) 95%
